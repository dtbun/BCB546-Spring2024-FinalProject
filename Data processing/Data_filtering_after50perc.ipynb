```python
with open("miRNA.txt") as file:                 #Using with to temporarily open file until task is complete
    lines = [line.rstrip() for line in file]    #Using rstrip to write the miRNA.txt line by line to lines object
    

only3p = [x for x in lines if "-3p" in x ]      #Take only the miRNA with -3p and place into object only3p
only5p = [x for x in lines if "-5p" in x ]      #Take only the miRNA with -5p and place into object only5p
not3p = [x for x in lines if "-3p" not in x ]   #Start filtering out and remove miRNA with -3p
not3p5p = [x for x in not3p if "-5p" not in x ] #Final filtering to remove all -3p and -5p miRNA and store into object

linesno3p = list(map(lambda x: x.replace('-3p', ''), only3p)) #make a list using map function and lambda function to remove -3p by replacing it with a blank
linesno5p = list(map(lambda x: x.replace('-5p', ''), only5p)) #make a list using map function and lambda function to remove -5p by replacing it with a blank
#map function allows us to work with iterables without a for loop
#lambda function is an anonymous function that lets us push output to a higher function. In this case to remove a specific string in their respective files

#Below are functions for printing out each list into a file, writing the list every newline. Not necessary as the objects have the correct items we need.

#with open('50perce_mirna_3p.txt','w') as fp: 
#    for item in linesno3p:
#        # write each item on a new line
#        fp.write("%s\n" % item)
#    print('Done')
#
#with open('50perce_mirna_5p.txt','w') as fp:
#    for item in linesno5p:
#        # write each item on a new line
#        fp.write("%s\n" % item)
#    print('Done')
#
#with open('50perce_mirna_nop.txt','w') as fp:
#    for item in not3p5p:
#        # write each item on a new line
#        fp.write("%s\n" % item)
#    print('Done')

```

```python
import pandas as pd #import pandas as pd for usage

common_elements_3p = pd.DataFrame(common_elements_3p) #using pandas to change to dataframe for data handling
common_elements_5p = pd.DataFrame(common_elements_5p) #using pandas to change to dataframe for data handling
common_elements_nop = pd.DataFrame(common_elements_nop) #using pandas to change to dataframe for data handling

append_3p = '-3p' #create object to store string we want to append
append_5p = '-5p' #create object to store string we want to append

common_elements_3p[0] = common_elements_3p[0].astype(str) + append_3p #convert our dataframe into str type, append our suffix to it, then save it back to the original dataframe at position 0
common_elements_5p[0] = common_elements_5p[0].astype(str) + append_5p #convert our dataframe into str type, append our suffix to it, then save it back to the original dataframe at position 0

togetha = pd.concat([common_elements_3p, common_elements_5p,common_elements_nop], ignore_index=True, sort=False) #concatenate all of the dataframes so all miRNA are back in one dataframe, ignore index to keep it aligned, no sort

#export cocatenated dataframe to a tsv file if desired
#togetha.to_csv('togetha.tsv', sep="\t") 
```
